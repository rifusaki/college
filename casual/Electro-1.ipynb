```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def f(X,Y,L,V,lim):
    term_1 = 0
    res = np.empty_like(X)
    for x in range(1,X.shape[0]):
        for y in range(1,Y.shape[1]):
            for n in range(1,lim):
                term_1 += (-1**n)*np.exp((-n)*np.pi*y/L)*((2*V)/(n*np.pi/L))*np.sin(n*np.pi*x/L)
            res[x,y] = term_1 + V*L/x
    return res
```

```python
L = 1
X, Y = np.meshgrid(np.linspace(0.1,L,100), np.linspace(0.1,L,100))
```

```python
plt.imshow(f(X,Y,L,10,100), cmap=cm.plasma, interpolation='bilinear')
#plt.imshow(Y, cmap=cm.plasma, interpolation='bilinear')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('$\Phi (x,y)$')
plt.show()
```

```python
plt.contourf(X,Y,f(X,Y,L,10,100), cmap=cm.plasma)
```
